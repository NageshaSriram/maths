# Vector and Matrix Operations 


4. Find the magnitude/ L1 and L2 norm of the vector [2,3] and state which has has a highest value.


```python
v = np.array([2, 3])
print(np.linalg.norm(v))
print("Manhattan L1", np.linalg.norm(v, ord=1))
print("Euclidean L2", np.linalg.norm(v, ord=2))
```

### Scalar and Vector Projections
7. Find the scalar and vector projections of the vector  u = 8i + 6j on vector v = i - 7j


```python
u = np.array([8, 6])
v = np.array([1, -7])
```

```python
# scalar
np.dot(u, v) / np.linalg.norm(v)
```

```python
scal = np.dot(v, u) / np.linalg.norm(v)
scal
```

```python
(scal * v) / np.linalg.norm(v)
```
