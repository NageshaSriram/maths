```python
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
import soundfile as sf
```

```python
# 1. Record your voice signal for 30 seconds (speak slowly and record)
duration = 30  # seconds
fs = 16000     # Sampling frequency (16 kHz is standard for speech)
print("Recording... Speak now")
voice = sd.rec(int(duration * fs), samplerate=fs, channels=1)
sd.wait()
print("Recording complete!")

# Save for playback
sf.write("voice.wav", voice, fs)

# Play recorded voice
print("Playing recorded voice...")
sd.play(voice, fs)
sd.wait()
```

```python
# 2. Print the shape of the signal
print("Shape of signal:", voice.shape)   # (num_samples, 1)
print("Is it 1D vector? ->", "Yes" if voice.ndim == 1 else "No, it is 2D with single channel")
```
