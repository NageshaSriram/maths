```python
import sympy as sp

# Define symbol and function
x = sp.Symbol('x')
f = x**2

# Gradient descent parameters
LR = 0.001 # Linear Regression
x_val = 5.0  # Random initial value
max_iter = 10000

for i in range(max_iter):
    grad = sp.diff(f, x).evalf(subs={x: x_val})
    step = LR * grad
    if abs(step) <= 0.0001:
        break
    x_val = x_val - step

print(f"Minimum at x ≈ {x_val:.5f}, f(x) ≈ {x_val**2:.5f}, iterations: {i+1}")
```
