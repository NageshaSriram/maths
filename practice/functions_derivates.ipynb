```python
cri_points = sp.solveset(f1, x, domain=sp.Reals)
cri_points
```

```python
f2.subs(x, cri_points.args[0])
```

```python
maxima, minima = [], []
for point in cri_points:
    concavity = f2.subs(x, point)
    if concavity < 0:
        maxima.append(point)
    elif concavity > 0:
        minima.append(point)
```

```python
# Define the variable
x = sp.symbols('x')

# Define the profit function
P = -8*x**2 + 3200*x - 80000

# Find the first derivative of P
P_prime = sp.diff(P, x)

# Solve for critical points (where derivative equals zero)
critical_points = sp.solve(P_prime, x)

# Find the second derivative to test concavity
P_double_prime = sp.diff(P_prime, x)

# Evaluate the second derivative at critical points
second_derivative_values = [P_double_prime.subs(x, cp) for cp in critical_points]

# Find the point that maximizes profit
max_profit_x = [cp for cp, sd in zip(critical_points, second_derivative_values) if sd < 0][0]

# Display the result
print(f"The number of apartments to rent for maximum profit is: {max_profit_x}")
```
