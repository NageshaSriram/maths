```python
import matplotlib.pyplot as plt
import numpy as np

# Define the vector
v = np.array([3, 2, 3, 4])

# Plot
plt.figure()
plt.axhline(0, color='gray', linewidth=0.5)  # x-axis
plt.axvline(0, color='gray', linewidth=0.5)  # y-axis
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue')

# Set limits
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Vector Plot")
plt.show()
```

```python
vectors = np.array([[3, 2], [1, 4], [-2, 1]])

plt.figure()
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)

for v in vectors:
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1)

plt.xlim(-3, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.title("Multiple Vectors")
plt.show()
```

```python
# Function to create square of side x
def plot_points(matrix, ls=':', lw=6):
    x_points, y_points = matrix
    size = len(x_points)
    colors = ['red', 'green', 'blue', 'black'] if not None else colors
    for i in range(size):
        plt.plot(x_points[i], y_points[i], color=colors[i], marker='*')
        plt.plot([x_points[i], y_points[i]], 
             [x_points[(i+1)%size], y_points[(i+1)%size]], 
             linestyle=ls, linewidth=lw)
```

```python
# Using above function to create a square of side 20
x_points = np.array([0, 0, 20, 20])
y_points = np.array([0, 20, 20, 0])
matrix = np.array([x_points, y_points])
colors = ['red', 'green', 'blue', 'black']
size = len(x_points)
plot_points(matrix)
plt.ylim([-25,45])
plt.xlim([-25,45])
#plt.axes().set_aspect('equal')
plt.show()
```

```python
# Define Transformations
def get_rotation(angle):
    angle = np.radians(angle)
    return np.array([

[np.cos(angle), -np.sin(angle), 0],
[np.sin(angle), np.cos(angle), 0],
])

res = get_rotation(90)
```
